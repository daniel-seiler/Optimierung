$$\vec{y}=
\begin{pmatrix}
1.5 \\
1.75 \\
2.5
\end{pmatrix}
, A=
\begin{pmatrix} 
    1 & 1 \\
    1.5 & 1 \\
    3 & 1 
\end{pmatrix}
$$

```python
import numpy as np

m = 0.5
c = 2
a = 0
b = 10
n_points = 50
noise_variance = 0.1
x = np.random.uniform(a, b, n_points)
y = m * x + c + np.random.normal(0, noise_variance, n_points)
```

```python
def create_matrices(data):
    A = np.column_stack((data[0], np.ones(len(data[0]))))
    y = data[1]
    return A, y

A, y_vec = create_matrices((x, y))
```

```python
# using qr-solver
Q, R = np.linalg.qr(A)
theta = np.linalg.solve(R, Q.T @ y_vec)
theta
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.axline((0, theta[1]), slope=theta[0], color="g")
plt.axline((0, c), slope=m, color="b")
plt.show()
```
